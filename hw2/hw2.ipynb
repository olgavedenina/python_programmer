```python
import json
import urllib.request

users = ['elmiram', 'maryszmary', 'lizaku', 'nevmenandr', 'ancatmara', 'roctbb', 'akutuzov', 'agricolamz', 'lehkost',
            'kylepjohnson', 'mikekestemont', 'demidovakatya', 'shwars', 'JelteF', 'timgraham',
            'arogozhnikov', 'jasny', 'bcongdon', 'whyisjake', 'gvanrossum']

def choice(users):  # choice out of the list
    choice = input('Введите имя пользователя Github: ')
    if choice in users:
        print('Вы выбрали пользователя %s.' % choice)
        return choice
    else:
        print('Такого пользавателя нет в списке. Введите имя пользователя еще раз.')
        return choice(users)


def getjson(url):  # json request
    response = urllib.request.urlopen(url)
    text = response.read().decode('utf-8')
    data = json.loads(text)
    return data


def repo_description(data):  # repositories with descriptions
    descript = dict()
    for i in data:
        descript[str(i['name'])] = str(i['description'])
    return descript


def language(data):  # languages
    ling = dict()
    for i in data:
        if i['language'] in ling.keys():
            ling[i['language']].append(i['name'])
        elif i['language'] is not None:
            ling[i['language']] = list()
            ling[i['language']].append(i['name'])
    return ling


def followers(data):  # amount of followers
    return len(data)


def most(item1, item2, name1, name2):  # comparison
    if item1 < item2:
        return item2, name2
    else:
        return item1, name1


def most_language(diction):  # the most popular language
    most = ''
    for lang in diction.keys():
        if most == '' or diction[most] < diction[lang]:
            most = lang
    return most


def ling_diction(newdict, l_d):  # language frequency
    for i in newdict:
        if i not in l_d.keys():
            l_d[i] = len(newdict[i])
        else:
            l_d[i] += len(newdict[i])
    return l_d


def main():    
    uchoice = choice(users)
    most_repo = 0
    most_follow = 0
    most_lang = ''
    langdict = dict()
    name_repo = ''
    name_follow = ''
    for user in users:
        repo_dt = getjson('https://api.github.com/users/%s/repos' % user)
        follow_dt = getjson('https://api.github.com/users/%s/followers' % user)
        follow = followers(follow_dt)
        descript = repo_description(repo_dt)
        ling = language(repo_dt)
        most_follow, name_follow = most(most_follow, follow, name_follow, user)
        most_repo, name_repo = most(most_repo, len(descript), name_repo, user)
        langdict = ling_diction(ling, langdict)
        most_lang = most_language(langdict)
        if user == uchoice:
            print('Его репозитории:')
            for d in descript:
                print('%s : %s' % (d, descript[d]))
            print('Он использует языки ', end='')
            for i in ling:
                print('на %s в ' % i, end='')
                for j in ling[i]:
                    print(' %s' % j, end='')
                print('\nвсего в %s' % len(ling[i]))
    print('Самый популярный язык %s' % most_lang)
    print('Больше всего репозиториев у %s' % name_repo)
    print('Больше всего подписчиков у %s' % name_follow)


if __name__ == '__main__':
    main()
```
